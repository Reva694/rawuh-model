```python
import cv2
from matplotlib import pyplot as plt

# Judul dan warna untuk tiap subplot
titles = ['Histogram Biru', 'Histogram Hijau', 'Histogram Merah']
colors = ('b', 'g', 'r')

plt.figure(figsize=(15, 4)) # Lebar agar 3 gambar muat ke samping

for i, col in enumerate(colors):
    plt.subplot(1, 3, i+1) # Membuat 1 baris, 3 kolom
    histr = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(histr, color=col)
    plt.title(titles[i])
    plt.xlim([0, 256])

plt.tight_layout() # Biar antar grafik gak tabrakan
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Load dan bersihkan noise sedikit
img = cv2.imread("/content/images (2).jpg", 0) # Load as grayscale
blur = cv2.GaussianBlur(img, (5,5), 0)

# 2. Reshape data untuk K-Means
data = blur.reshape((-1,1))
data = np.float32(data)

# 3. K-Means (K=3 untuk memisahkan background, jaringan, dan tumor)
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 3
ret, label, center = cv2.kmeans(data, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

# 4. Ambil cluster dengan nilai intensitas tertinggi (Tumor biasanya paling terang)
center = np.uint8(center)
tumor_cluster = np.argmax(center) # Mencari index cluster yang paling putih
res = center[label.flatten()]
res2 = res.reshape((img.shape))

# Buat mask hanya untuk cluster tumor
mask = np.zeros_like(img)
mask[label.reshape(img.shape) == tumor_cluster] = 255

# 5. Pembersihan Akhir (Morphology) untuk membuang bintik kecil
kernel = np.ones((5,5), np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

# Visualisasi
plt.figure(figsize=(12, 4))
plt.subplot(131); plt.title("Original Grayscale"); plt.imshow(img, cmap='gray')
plt.subplot(132); plt.title("K-Means Mask (Tumor Only)"); plt.imshow(mask, cmap='gray')
plt.subplot(133); plt.title("Hasil Ekstraksi"); plt.imshow(cv2.bitwise_and(img, img, mask=mask), cmap='gray')
plt.show()
```
