```python
# MobileNetV2 fast pipeline
# ---------------------------------------------------------------
# 1) Mount Drive
# from google.colab import drive
# drive.mount('/content/drive')

import os, math, itertools
import numpy as np
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

# -------------------------
# Paths
# -------------------------
train_dir = "/content/drive/MyDrive/Dataset_Sistem_Presensi_Wajah_V2.0/Data Train"
test_dir  = "/content/drive/MyDrive/Dataset_Sistem_Presensi_Wajah_V2.0/Data Test"

# -------------------------
# Config
# -------------------------
IMG_SIZE = (96,96)
BATCH_SIZE = 64
EPOCHS = 18
LR = 1e-4

AUTOTUNE = tf.data.AUTOTUNE

# -------------------------
# Helper: get label from filename (ambil NIM sebelum underscore)
# -------------------------
def get_label_from_filename(fname):
    return os.path.basename(fname).split("_")[0]

# -------------------------
# 1. Build file lists & labels (from flat folder)
# -------------------------
def list_files_and_labels(folder):
    files = [os.path.join(folder,f) for f in sorted(os.listdir(folder)) if f.lower().endswith((".jpg",".jpeg",".png"))]
    labels = [get_label_from_filename(f) for f in files]
    return np.array(files), np.array(labels)

train_files, train_labels_text = list_files_and_labels(train_dir)
test_files,  test_labels_text  = list_files_and_labels(test_dir)

print("Train files:", train_files.shape, "Test files:", test_files.shape)

# -------------------------
# Label encode text -> integer
# -------------------------
le = LabelEncoder()
le.fit(np.concatenate([train_labels_text, test_labels_text]))  # pastikan mapping sama
y_train_int = le.transform(train_labels_text)
y_test_int  = le.transform(test_labels_text)
NUM_CLASSES = len(le.classes_)
print("Jumlah kelas:", NUM_CLASSES)

# Save classes for later use (e.g., loading in inference)
np.save("/content/drive/MyDrive/labels.npy", le.classes_)
print("Labels saved to: /content/drive/MyDrive/labels.npy")
# -------------------------
# 2. TF dataset pipeline (fast)
# -------------------------
def preprocess_image(path, label):
    # read file, decode, resize, convert float32 and MobileNetV2 preprocess
    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, IMG_SIZE)
    image = tf.cast(image, tf.float32) / 255.0
    # MobileNetV2 expects inputs in [-1,1]
    image = (image - 0.5) * 2.0
    return image, label

def make_dataset(files, labels, training=True):
    ds = tf.data.Dataset.from_tensor_slices((files, labels))
    if training:
        ds = ds.shuffle(buffer_size=len(files), seed=42)
    ds = ds.map(preprocess_image, num_parallel_calls=AUTOTUNE)
    if training:
        ds = ds.batch(BATCH_SIZE).cache().prefetch(AUTOTUNE)
    else:
        ds = ds.batch(BATCH_SIZE).cache().prefetch(AUTOTUNE)
    return ds

train_ds = make_dataset(train_files, y_train_int, training=True)
test_ds  = make_dataset(test_files,  y_test_int,  training=False)

# -------------------------
# 3. Build MobileNetV2 model (feature extractor frozen)
# -------------------------
base = MobileNetV2(weights='imagenet', include_top=False, input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3))
base.trainable = False  # freeze for speed

x = base.output
x = GlobalAveragePooling2D()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.35)(x)
out = Dense(NUM_CLASSES, activation='softmax')(x)

model = Model(inputs=base.input, outputs=out)
opt = Adam(learning_rate=LR)
model.compile(optimizer=opt, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# -------------------------
# 4. Train (fast)
# -------------------------
history = model.fit(train_ds, validation_data=test_ds, epochs=EPOCHS, verbose=1)

# -------------------------
# 5. Save model (optional)
# -------------------------
save_path = "/content/drive/MyDrive/FaceRec_MobileNetV2_fast.h5"
model.save(save_path)
print("Model saved to:", save_path)

# -------------------------
# 6. Evaluate: print numeric accuracies
# -------------------------
train_loss, train_acc = model.evaluate(train_ds, verbose=0)
test_loss, test_acc   = model.evaluate(test_ds, verbose=0)
print(f"\nAkurasi Training : {train_acc*100:.2f}%")
print(f"Akurasi Testing  : {test_acc*100:.2f}%")

# -------------------------
# 7. Plot accuracy & loss (rapi)
# -------------------------
plt.figure(figsize=(8,5))
plt.plot(history.history['accuracy'], marker='o', label='Train accuracy', linewidth=2)
plt.plot(history.history['val_accuracy'], marker='o', label='Val accuracy', linewidth=2)
plt.title("Accuracy per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.ylim(0,1.01)
plt.grid(alpha=0.25)
plt.legend()
plt.show()

plt.figure(figsize=(8,5))
plt.plot(history.history['loss'], marker='o', label='Train loss', linewidth=2)
plt.plot(history.history['val_loss'], marker='o', label='Val loss', linewidth=2)
plt.title("Loss per Epoch")
plt.xlabel("Epoch")
plt.grid(alpha=0.25)
plt.legend()
plt.show()

# -------------------------
# 8. Predictions on test set (collect probs & preds)
# -------------------------
# we need flattened arrays aligned with filepaths order
y_true = []
y_pred = []
y_prob = []

for images, labels in test_ds:
    probs = model.predict(images, verbose=0)
    preds = np.argmax(probs, axis=1)
    y_prob.append(probs)
    y_pred.append(preds)
    y_true.append(labels.numpy())

y_prob = np.vstack(y_prob)
y_pred = np.concatenate(y_pred)
y_true = np.concatenate(y_true)

# -------------------------
# 9. Classification report (printed)
# -------------------------
print("\n=== Classification report (some rows) ===")
print(classification_report(y_true, y_pred, target_names=le.classes_, zero_division=0))

# -------------------------
# 10. Confusion matrix (big, no labels)
# -------------------------
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6,6))
# plot matrix without tick labels to avoid clutter
plt.imshow(cm, cmap='Blues', aspect='auto')
plt.colorbar()
plt.title(f"Confusion Matrix (no labels) — Train {train_acc*100:.1f}% / Test {test_acc*100:.1f}%")
plt.xticks([]); plt.yticks([])
plt.show()

# -------------------------
# 11. Top-3 hardest to predict (by lowest max confidence) + show 3 mispred per person
# -------------------------
confidences = np.max(y_prob, axis=1)
lowest_idxs = np.argsort(confidences)[:30]  # consider many lowest to find distinct persons
# compute misclassification counts per class
miscounts = {}
for t,p,i in zip(y_true, y_pred, range(len(y_true))):
    if t != p:
        key = le.classes_[t]
        miscounts[key] = miscounts.get(key, 0) + 1

# get top3 most-misclassified by count (fall back to lowest confidence if tie)
top3_by_mis = sorted(miscounts.items(), key=lambda x: x[1], reverse=True)[:3]
top3_names = [name for name,_ in top3_by_mis]

# If miscounts insufficient (rare), fallback to lowest confidence persons
if len(top3_names) < 3:
    low_persons = []
    for idx in lowest_idxs:
        low_persons.append(le.classes_[y_true[idx]])
    # unique preserve order
    seen=set()
    for p in low_persons:
        if p not in seen:
            seen.add(p)
            top3_names.append(p)
        if len(top3_names)==3:
            break

print("\nTop-3 orang paling sulit dikenali (nama/NIM) & jumlah salah prediksi (jika ada):")
for name in top3_names:
    print("-", name, "->", miscounts.get(name,0), "salah")

# show up to 3 mispredicted images (or lowest-confidence ones) per person with simple reasons
def image_brightness(img_path):
    img = cv2.imread(img_path)
    if img is None: return None
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    return float(np.mean(hsv[:,:,2]))

def image_blur_score(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if img is None: return None
    return float(cv2.Laplacian(img, cv2.CV_64F).var())

filepaths = np.array(test_files)  # order is the same as test_files used for dataset
# gather indices of mispred for each top person
for person in top3_names:
    person_idx = np.where(le.classes_ == person)[0][0]
    mis_indices = [i for i,(t,p) in enumerate(zip(y_true, y_pred)) if (t==person_idx and t!=p)]
    print(f"\n--- Contoh (max 3) salah prediksi untuk {person} (total salah {len(mis_indices)}) ---")
    if len(mis_indices)==0:
        # no mispred -> show lowest confidence images of that person
        person_idxs = [i for i,t in enumerate(y_true) if t==person_idx]
        # choose lowest confidence among them
        person_sorted = sorted(person_idxs, key=lambda i: confidences[i])[:3]
        show_idxs = person_sorted
    else:
        show_idxs = mis_indices[:3]

    plt.figure(figsize=(12,3))
    for k,idx in enumerate(show_idxs):
        img_path = filepaths[idx]
        brightness = image_brightness(img_path)
        blur = image_blur_score(img_path)
        reasons = []
        if brightness is not None:
            if brightness < 60:
                reasons.append("gelap")
            elif brightness > 200:
                reasons.append("overexposed")
            else:
                reasons.append("pencahayaan OK")
        if blur is not None:
            if blur < 100:
                reasons.append("buram")
            else:
                reasons.append("tajam")
        # jumlah data train for this person
        train_person_dir = None
        # we can count by scanning train_files
        train_count = np.sum([1 for lab in train_labels_text if lab==person])
        if train_count < 20:
            reasons.append("sedikit data train")
        else:
            reasons.append("data train cukup")

        ax = plt.subplot(1, len(show_idxs), k+1)
        img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
        ax.imshow(img)
        ax.axis('off')
        pred_name = le.classes_[y_pred[idx]]
        title = f"True: {person}\nPred: {pred_name}\nConf: {confidences[idx]:.2f}\nReason: {', '.join(reasons)}"
        ax.set_title(title, fontsize=9)
    plt.show()

# -------------------------
# 12. (Optional) show one uploaded image from session (developer-provided file path)
# -------------------------
# using file path from conversation history (example)
uploaded_example = "/mnt/data/9ffbf667-f6bf-405a-b322-b445fceab891.png"
if os.path.exists(uploaded_example):
    from IPython.display import display, Image
    print("\nContoh file yang ada di session (ditampilkan):", uploaded_example)
    display(Image(uploaded_example, width=300))
else:
    print("\nTidak menemukan contoh uploaded file pada path:", uploaded_example)

# -------------------------
# 13. Short recommendations (printed)
# -------------------------
print("\nRekomendasi singkat:")
print("- Jika ingin lebih tinggi lagi, lakukan fine-tune: unfreeze 5-10 layer terakhir base MobileNetV2 dan train 5-8 epoch dengan LR kecil (1e-5).")
print("- Jika ingin training lebih cepat lagi, aktifkan mixed precision dan pastikan GPU runtime.")
print("- Preprocess: face-crop + align biasanya menambah 3-7% akurasi.")
```

```python
import os
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from sklearn.metrics.pairwise import cosine_similarity

# Path ke data training
train_dir = "/content/drive/MyDrive/Dataset_Sistem_Presensi_Wajah_V2.0/Data Train"

# Load training faces
def load_images(folder):
    images, names = [], []
    for f in os.listdir(folder):
        if f.lower().endswith(('.jpg', '.jpeg', '.png')):
            img = load_img(os.path.join(folder, f), target_size=(224,224))
            arr = img_to_array(img)
            arr = preprocess_input(arr)
            images.append(arr)
            names.append(f)
    return np.array(images), names

X_train, train_names = load_images(train_dir)
print(f"Jumlah data train: {len(X_train)}")

# Ekstraksi fitur wajah (menggunakan MobileNetV2)
model = MobileNetV2(weights='imagenet', include_top=False, pooling='avg')
train_features = model.predict(X_train, verbose=1)

# Ekstraksi fitur foto realtime
img_resized = cv2.resize(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), (224,224))
img_arr = np.expand_dims(preprocess_input(img_to_array(img_resized)), axis=0)
test_feature = model.predict(img_arr)

# Cek kemiripan
sim = cosine_similarity(test_feature, train_features)[0]
best_idx = np.argmax(sim)
best_score = sim[best_idx]
best_match = train_names[best_idx]

print(f"\nHasil Pengenalan Wajah:")
print(f" Terdekat dengan: {best_match}")
print(f" Skor kemiripan: {best_score:.3f}")
```

```python
# Kode untuk Face Detection dan Face Recognition pada foto yang diambil
import cv2
import numpy as np
import tensorflow as tf
import pickle
import matplotlib.pyplot as plt

# Tambahan impor untuk take_photo function
from IPython.display import display, Javascript, Image
from google.colab.output import eval_js
import base64

# Definisi fungsi js_to_image
def js_to_image(js_reply):
    """Konversi JavaScript base64 ke NumPy array (OpenCV)"""
    image_bytes = base64.b64decode(js_reply.split(',')[1])
    jpg_as_np = np.frombuffer(image_bytes, dtype=np.uint8)
    img = cv2.imdecode(jpg_as_np, flags=1)
    return img

# Definisi fungsi take_photo (dipindahkan dari Z_uYq6yJOGl9)
def take_photo(filename='photo.jpg', quality=0.8):
    """Menampilkan kamera dan mengambil 1 foto"""
    js = Javascript('''
      async function takePhoto(quality) {
        const div = document.createElement('div');
        const capture = document.createElement('button');
        capture.textContent = ' Capture';
        div.appendChild(capture);

        const video = document.createElement('video');
        video.style.display = 'block';
        const stream = await navigator.mediaDevices.getUserMedia({video: true});
        document.body.appendChild(div);
        div.appendChild(video);
        video.srcObject = stream;
        await video.play();

        // Ukuran video
        google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

        // Tunggu tombol di-klik
        await new Promise((resolve) => capture.onclick = resolve);

        const canvas = document.createElement('canvas');
        canvas.width = video.videoWidth;
        canvas.height = video.videoHeight;
        canvas.getContext('2d').drawImage(video, 0, 0);
        stream.getVideoTracks()[0].stop();
        div.remove();

        const dataUrl = canvas.toDataURL('image/jpeg', quality);
        return dataUrl;
      }
      ''')
    display(js)
    data = eval_js('takePhoto({})'.format(quality))
    img = js_to_image(data)
    cv2.imwrite(filename, img)
    print(f"Foto disimpan ke {filename}")
    return filename, img

# Asumsi `preprocess_face` sudah didefinisikan di sel sebelumnya.
# (preprocess_face dari f6iEEqPRtq1W)

# --------------------------
# Load model & label encoder
# --------------------------
model = tf.keras.models.load_model("/content/drive/MyDrive/FaceRec_MobileNetV2_fast.h5")

with open('/content/drive/MyDrive/label_encoder.pkl', 'rb') as f:
    le = pickle.load(f)

# --------------------------
# Load face detector
# --------------------------
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# --------------------------
# Capture photo and perform recognition
# --------------------------
print("Mengambil foto dari kamera untuk deteksi dan pengenalan wajah...")
# Panggil fungsi take_photo yang sudah didefinisikan di sel lain
filename, img = take_photo(filename='captured_face.jpg')

# Konversi gambar yang ditangkap dari BGR (default OpenCV) ke RGB untuk matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Lakukan deteksi wajah pada gambar
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.3, 5)

# Jika wajah ditemukan, proses wajah pertama
if len(faces) > 0:
    (x, y, w, h) = faces[0] # Ambil wajah pertama yang terdeteksi

    face_crop = img[y:y+h, x:x+w]

    # Preprocess wajah untuk model
    face_processed = preprocess_face(face_crop)

    # Prediksi menggunakan model
    pred = model.predict(face_processed, verbose=0)[0] # verbose=0 untuk menghindari progress bar
    class_id = np.argmax(pred)
    confidence = pred[class_id]

    # Dapatkan nama orang dari label encoder
    name = le.classes_[class_id]

    # Gambar persegi panjang dan teks hasil pengenalan pada gambar RGB
    cv2.rectangle(img_rgb, (x, y), (x+w, y+h), (0, 255, 0), 2)
    cv2.putText(img_rgb, f"{name} ({confidence*100:.1f}%)",
                (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    print(f"Wajah terdeteksi: {name} dengan confidence {confidence*100:.1f}%")
else:
    print("Tidak ada wajah yang terdeteksi pada foto yang diambil.")

# Tampilkan gambar dengan hasil deteksi/pengenalan
plt.figure(figsize=(8, 6))
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Hasil Pengenalan Wajah")
plt.show()
```
